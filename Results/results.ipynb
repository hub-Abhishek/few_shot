```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('few_shot_untrained_scores_complete_df2.csv')
```

```python
fig, axs = plt.subplots(2, figsize=(10, 10), sharex=False, sharey=True)

axs[0].set_title('Accuracy vs n')
axs[0].plot(results.n_way.unique(), results.groupby('n_way').score.mean(), label='Trained model')
axs[0].plot(results.n_way.unique(), results.groupby('n_way').untrained_score.mean(), label='VGG19 with imagenet')
axs[0].set_xlabel('N - way')
axs[0].set_ylabel('Accuracy')
axs[0].legend()

axs[1].set_title('Accuracy vs k')
axs[1].plot(results.k_shot.unique(), results.groupby('k_shot').score.mean(), label='Trained model')
axs[1].plot(results.k_shot.unique(), results.groupby('k_shot').untrained_score.mean(), label='VGG19 with imagenet')
axs[1].set_xlabel('k - shot')
axs[1].set_ylabel('Accuracy')
axs[1].legend(loc=4)

fig.savefig('metrics.png')
```
